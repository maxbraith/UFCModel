```python
#imports
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ufc_fighters_statistics2024.csv')
dfAlt = pd.read_csv('alt_fighter_stats2024.csv')

df.head()
```

```python
dfAlt.head()
```

```python
print(len(df))
print(len(dfAlt))
```

```python
#clean nicknames in dfAlt
dfAlt['nickname'] = dfAlt['nickname'].str.replace('"', '')
```

```python
#giving nans characters so dups with nans for nickname cannot match
df['nickname'].fillna('-', inplace=True)
dfAlt['nickname'].fillna('@', inplace=True)
```

```python
#outer merge the dfs
dfMerged = pd.merge(df, dfAlt, on=['name', 'nickname'], how='outer', indicator=True)
dfMerged.head()
```

```python
len(dfMerged)
```

```python
#query dfMerged for intersections
dfMatches = pd.DataFrame(dfMerged.query('_merge == "both"'))
dfMatches.head()
```

```python
print(len(dfMatches))
```

```python
#query dfMerged for symmetric difference
notMatched = pd.DataFrame(dfMerged.query('_merge != "both"'))
notMatched.head()
```

```python
#find the nonMatched items from dfAlt
dfAltNoMatch = pd.merge(notMatched, dfAlt, on=['name', 'nickname'], how='inner')
print(len(dfAltNoMatch))
```

```python
#find the nonMatched items from df
dfNoMatch = pd.merge(notMatched, df, on=['name', 'nickname'], how='inner')
print(len(dfNoMatch))
```

```python
#writing non matched instances of both dfs to csv for manual analysis
dfNoMatch.to_csv('SymmetricDifferenceFighterStats.csv', index=False)
dfAltNoMatch.to_csv('SymmetricDifferenceAlt.csv', index=False)
```
