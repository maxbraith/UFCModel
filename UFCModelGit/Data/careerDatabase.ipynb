```python
#assign nicknames for matching names found withink dataframes
count = 1
for index, row in df.iterrows():
    name = row['name']
    nickname = row['nickname']
    dup_check = df.loc[df['name'] == name]
    dup_check2 = dfAlt.loc[dfAlt['name'] == name]
    if(len(dup_check == 1) and len(dup_check2) == 1):
        try:
            altIndex = dfAlt.loc[dfAlt['name'] == name].index[0]
            altNickname = dfAlt.loc[altIndex, 'nickname']
            if(nickname != altNickname):
                if(nickname == '!' and altNickname == '~'):
                    df.loc[index, 'nickname'] = str(count)
                    dfAlt.loc[altIndex, 'nickname'] = str(count)
                    count+=1
                elif(nickname != '!'):
                    dfAlt.loc[altIndex, 'nickname'] = nickname
                elif(altNickname != '~'):
                    df.loc[index, 'nickname'] = altNickname
        except:
            pass

        
```

```python
#outer merge the dfs
dfMerged = pd.merge(df, dfAlt, on=['name', 'nickname'], how='outer', indicator=True)
dfMerged.head()
```

```python
len(dfMerged)
```

```python
#query dfMerged for intersections
dfMatches = pd.DataFrame(dfMerged.query('_merge == "both"'))
dfMatches.head()
```

```python
print(len(dfMatches))
```

```python
#query dfMerged for symmetric difference
notMatched = pd.DataFrame(dfMerged.query('_merge != "both"'))
notMatched.head()
```

```python
#find the nonMatched items from dfAlt
dfAltNoMatch = pd.merge(notMatched, dfAlt, on=['name', 'nickname'], how='inner')
print(len(dfAltNoMatch))
```

```python
#find the nonMatched items from df
dfNoMatch = pd.merge(notMatched, df, on=['name', 'nickname'], how='inner')
print(len(dfNoMatch))
```

```python
#writing non matched instances of both dfs to csv for manual analysis
dfNoMatch.to_csv('SymmetricDifferenceFighterStats.csv', index=False)
dfAltNoMatch.to_csv('SymmetricDifferenceAlt.csv', index=False)
```
